```python
 def try_Help_All_Package(package):

  x = dir(package)

  clist = []


  for inick in x:
    clist.append(str(package)+"."+inick)

#  return clist

  clist = ('[%s]' % ', '.join(map(str, clist)))
    
  for item in clist:
    help(item)

try_Help_All_Package("BeautifulSoup")
```
